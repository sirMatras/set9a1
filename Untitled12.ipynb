```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("results.csv")

data['time_us'] = data['duration_ms'] * 1000

summary = data.groupby(['type', 'algo']).agg({
    'time_us': ['mean', 'std', 'min', 'max'],
    'comparisons': ['mean', 'std', 'min', 'max']
})

summary.columns = ['_'.join(col) for col in summary.columns]
summary = summary.reset_index()

types = data['type'].unique()
algos = data['algo'].unique()

for t in types:
    plt.figure(figsize=(8, 5))
    sub = data[data['type'] == t]
    for algo in algos:
        d = sub[sub['algo'] == algo]
        if d.empty:
            continue
        plt.plot(d['size'], d['time_us'], marker='o', label=algo)
    plt.title(f"Time vs n (type = '{t}')")
    plt.xlabel('n (size of data)')
    plt.ylabel('Time (μs, log scale)')
    plt.yscale('log')
    plt.legend()
    plt.grid(True, which='both', ls='--')
    plt.tight_layout()
    plt.show()

for t in types:
    plt.figure(figsize=(8, 5))
    sub = data[data['type'] == t]
    for algo in algos:
        d = sub[sub['algo'] == algo]
        if d.empty:
            continue
        plt.plot(d['size'], d['comparisons'], marker='o', label=algo)
    plt.title(f"Char comparisons vs n (type = '{t}')")
    plt.xlabel('n (size of data)')
    plt.ylabel('Char comparisons')
    plt.legend()
    plt.grid(True, ls='--')
    plt.tight_layout()
    plt.show()

print("Empirical time complexity (time ∼ n^α):")
for t in types:
    print(f"Type = '{t}':")
    sub = data[data['type'] == t]
    for algo in algos:
        d = sub[sub['algo'] == algo].copy()
        d = d[d['time_us'] > 0]
        if len(d) < 2:
            continue
        x = np.log(d['size'])
        y = np.log(d['time_us'])
        alpha = np.polyfit(x, y, 1)[0]
        print(f"  {algo:15s}: α ≈ {alpha:.2f}")
    print()
```
